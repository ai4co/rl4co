```python
# define instance and solution path
name = "C101"
path_instances = "data/instances/"
path_solutions = "data/solutions/"
```

```python
# download from vrplib
vrplib.download_instance(name=name, path=path_instances)
vrplib.download_solution(name=name, path=path_solutions)
```

```python
# Read Solomon formatted instances
# .txt files are Solomon formatted, .vrp files are vrplib formatted
instance = vrplib.read_instance(path=f"{path_instances}{name}.txt", instance_format="solomon")
solution = vrplib.read_solution(path=f"{path_solutions}{name}.sol")
```
